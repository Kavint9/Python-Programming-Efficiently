## Mastering Python loops

**March 2020 update**:

* at 04:00 in the video: `geopy.geocoders.Nominatim` should be called with a unique `user_agent` to avoid user-policy violations. In addition, on a slow connection it may be necessary to give `Nominatim` a higher timeout (for example, `timeout = 10`).
* at 05:20 in the video: the `Nominatim` geocoder now returns Italy and Greece locations for Rome and Athens instead of the smaller U.S. towns, so the more detailed queries described in the video are not necessary.

```python
cities, years = [], []
for game in open('games.txt'):
    city = game.split()[:-1]
    cities.append(' '.join(city))
    year = game.split()[-1].strip('()')
    years.append(year)
print(cities)
print(len(years))
```

```python
geolocator = geopy.geocoders.Nominatim(user_agent = "test")
locations = {}
for city in cities:
    print("Locating",city)
    locations[city] = geolocator.geocode(city.split('/')[0])

print(locations)
```

```python
for city, loc in locations.items():
     print(city)
     print(loc.longitude, loc.latitude)#, loc.latitude)
```

```python
pp.figure(figsize=(11,5))
world = Basemap()
world.drawcoastlines(linewidth=1)
world.drawcountries(linewidth=0.25)

for city, loc in locations.items():
    world.plot(loc.longitude, loc.latitude,'r.', markersize=10,latlon=True)
    
```

```python
locations['Rome'] = geolocator.geocode('Rome, Italy')
locations['Athens'] = geolocator.geocode('Athens, greece')
```

```python
for city in sorted(cities[10:-1], key=len): # sort by length default by alph order
    print(city)
```

```python
for year, city in zip(years[:10], cities): # ends when the shorter of the two arrays end
    print(year, city)
```

```python
for city in enumerate(reversed(cities[:5])): #enum gives value and ordinal number
    print(city)

for i, city in enumerate(reversed(cities[:5])): #i is the ordinal number
    print(i, city)
```
