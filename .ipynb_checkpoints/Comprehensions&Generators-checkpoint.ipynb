### Code and data needed from previous videos

List of olympic cities and years:

```python
cities = []
years = []

for game in open('games.txt','r'):
    words = game.split()
    
    city = ' '.join(words[:-1])
    year = words[-1].strip('()')

    cities.append(city)
    years.append(year)
```

Geolocated olympic-city coordinates (from a JSON file):

```python
coordinates_by_city = json.load(open('coords.json','r'))
coordinates_by_city
```

## Comprehensions and generators

Comprehensions provide a way to allow us to generate  list, dict etc. from loops and conditions

```python
post_ww2 = [city + ':' + year for city, year in zip(cities, years) if int(year) > 1945]
```

```python
post_ww2
```

```python
# above as a dict
post_ww2_years = {city: int(year) for city, year in zip(cities, years) if int(year) > 1945}
post_ww2_years
```

*Sets:* unordered unidexed (=> cannot be accessed using indexes)collection {}

```python
post_1930 = {city for city, year in post_ww2_years.items() if int(year) > 1930}
post_1930
```

```python
pp.figure(figsize=(8,8))
world = Basemap(projection='ortho', lat_0 = 75, lon_0=0) #co ord in europe
world.drawcoastlines(linewidth=0.75)
world.drawcountries(linewidth=0.75)
for city, year in post_ww2_years.items():
    x,y = world(*coordinates_by_city[city])
    world.plot(x,y, 'r.')
    pp.text(x,y, year, fontsize=8, ha='center', va='bottom', color='navy')
```

 **Generators**: 
 > a function that can be iterated over, each call runs the function till the yield statement is reached.
 > Next call resumes operation till yield
 > __next__() works with generators
 
 E.g. Fibonacci as a generator function
 

```python
def fibonacci():
    f1, f2 = 0, 1
    while True:
        yield f2
        f1, f2 = f2, f1 + f2 # Does f1 get assigned before f2? if so won't f2 = f2 + f2 => no simulataneosly

f = fibonacci()              # if we directly pass fibonacci() - the function starts from the beginning => values init
[next(f) for i in range(15)] # next(<gen>) = <gen>.__next__()
```
